```python
signal = TriangleSignal(300)
duration = signal.period*3
segment = signal.make_wave(duration, framerate=10000)
wave = signal.make_wave(duration=0.01, framerate=10000)
segment.plot()
decorate(xlabel='Time (s)')
```

```python
spectrum = wave.make_spectrum()
spectrum.plot()
decorate(xlabel='Frequency (Hz)')
```

```python
spectrum.hs[0]
```

```python
spectrum.hs[0]=100
spectrum.plot()
decorate(xlabel='Frequency (Hz)')
```

```python
def f5(spectrum):
    spectrum.hs[0]=0
    for i in range(1,len(spectrum.fs)):
        spectrum.hs[i]/=spectrum.fs[i]
    return spectrum
```

```python
# 方波验证
signal = SquareSignal(1100)
duration = signal.period*4
segment = signal.make_wave(duration, framerate=10000)
wave = signal.make_wave(duration=1, framerate=10000)
spectrum = wave.make_spectrum()
spectrum.plot()
decorate(xlabel='Frequency (Hz)')
```

```python
wave.apodize()
wave.make_audio()
```

```python
spectrum = f5(spectrum)
spectrum.plot()
decorate(xlabel='Frequency (Hz)')
```

```python
wave = spectrum.make_wave()
wave.apodize()
wave.make_audio()
```
